```python
import os
import pandas as pd
```

```python
directori_execucio = os.getcwd()
ruta_carpeta_silver = os.path.join(directori_execucio, '2-Silver')
productes = [item for item in os.listdir(ruta_carpeta_silver) if os.path.isfile(os.path.join(ruta_carpeta_silver, item))]
ruta_csv_productes = os.path.join(ruta_carpeta_silver, productes[0])
df = pd.read_csv(ruta_csv_productes)
df.head()
```

```python
df = df[['display_name', 'categoria_1_name', 'categoria_3_name', 'bulk_price', 'unit_price', 'total_units', 'size_format', 'packaging', 'ingested_day', 'iva', 'is_pack']]
df.head()
```

```python
# Veiem com hi ha noms molt llargs, per tant els reduirem.
max(df['display_name'], key=len)
```

```python
# Fem un nom curt per no ocupar gran part del tiquet:
def short_name(cell_value, n=18):
    words = len(cell_value.split())
    chars = len(cell_value)

    if words <= 1 or chars <= n:
        return cell_value

    words = cell_value[0:n].split(' ')[0:-1] # La última fora perque pot quedar la paraula tallada
    return ' '.join([word for word in words if len(word) > 2])

df['display_name'] = df['display_name'].apply(short_name)
df.head()
```

```python
df['display_name'] = df['display_name'].str.upper()
df.head()
```

```python
# Calcular preus totals:

def calculate_price(row):
    if row['size_format'] == 'kg':
        # Verificar que category sea solo una palabra y que units sea NaN
        if ((row['categoria_3_name'] in ['Fruita i verdura', 'Verdura']) and (len(row['display_name'].split()) <= 2)):
            return pd.Series({'price': row['bulk_price'], 'size_format': 'kg'})
        else:
            return pd.Series({'price': row['unit_price'], 'size_format': 'ud'})

    elif row['size_format'] == 'ud':
        if not pd.isna(row['total_units']):
            return pd.Series({'price': row['unit_price'] * row['total_units'], 'size_format': 'ud'})
        else:
            return pd.Series({'price': row['unit_price'], 'size_format': 'ud'})

    else:
        return pd.Series({'price': row['unit_price'], 'size_format': 'ud'})

df[['price', 'size_format']] = df.apply(calculate_price, axis=1)

df.head()
```

```python
df.rename(columns={
    'display_name': 'name',
    'size_format': 'format',
    'categoria_3_name': 'subcategory',
    'categoria_1_name': 'category'
    
}, inplace=True)

df = df[['name', 'category', 'subcategory', 'price', 'iva', 'format', 'ingested_day']]

df.head()
```

```python
# Guardar el dataset tractat a la capa gold:
ruta_carpeta_gold = os.path.join(directori_execucio, '3-Gold')
ruta_csv_gold = os.path.join(ruta_carpeta_gold, 'productes_mercadona.csv')
df.to_csv(ruta_csv_gold, index=False)
```
