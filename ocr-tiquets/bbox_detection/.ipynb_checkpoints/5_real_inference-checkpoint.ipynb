```python
from ultralytics import YOLO
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
import ast
import seaborn as sns
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
from scipy.optimize import linear_sum_assignment
import os
from pdf2image import convert_from_path
import shutil
```

#### Llegim els pdf de tiquets reals i els convertim a imatge

```python
path = '../../generador-tiquets/exemples_tiquets/mercadona/'

pdfs = archivos_pdf = [f for f in os.listdir(path) if f.endswith('.pdf') and os.path.isfile(os.path.join(path, f))]
pdfs
```

#### Convertim aquests pdf a img

```python
image_root = "./images_from_pdf"

if os.path.exists(image_root):
    shutil.rmtree(image_root)
os.makedirs(image_root)
```

```python
for pdf in pdfs:
    ruta_pdf = os.path.join(path, pdf)
    pages = convert_from_path(ruta_pdf)
    pages[0].save(f"./{image_root}/{pdf.split('.')[0]}.png")
```

```python
images = [f for f in os.listdir(image_root) if f.endswith('.png')]

def number(nombre_archivo):
    parts = nombre_archivo.split('_')
    num = int(parts[1].split('.')[0])
    return num

images = sorted(images, key=number)
images
```

#### Mostrar imatges

```python
for ticket_id in images:
    pilImg = Image.open(f"{image_root}/{ticket_id}")
    print(pilImg.size)

    plt.imshow(pilImg)
    plt.axis('off');
    plt.show()
```

```python
base_width = 500

for ticket_id in images:
    img = Image.open(f"{image_root}/{ticket_id}")

    wpercent = (base_width / float(img.size[0]))
    hsize = int((float(img.size[1]) * float(wpercent)))
    img = img.resize((base_width, hsize), Image.Resampling.LANCZOS)
    img.save(f"{image_root}/resized_{ticket_id}")
```

```python
images_resized = [f for f in os.listdir(image_root) if f.startswith('resized_') and f.endswith('.png')]

def number(nombre_archivo):
    parts = nombre_archivo.split('_')
    num = int(parts[2].split('.')[0])
    return num

images_resized = sorted(images_resized, key=number)
images_resized
```

```python
for ticket_id in images_resized:
    pilImg = Image.open(f"{image_root}/{ticket_id}")
    print(pilImg.size)

    plt.imshow(pilImg)
    plt.axis('off');
    plt.show()
```

#### Fem inferencia amb les imatges provinents del pdf
- Hem fet un nou entrenament amb més imatges i epoch
- 15_000 imatges de train
- 50 epoch

```python
model = YOLO('./runs/detect/train_5k_15e_v3/weights/best.pt')

inference = model(image_root)
```

```python
all_images = images_resized + images
all_images
```

```python
image_root = '../../generador-tiquets/exemples_tiquets/lidl/'

lidl_images = [f for f in os.listdir(image_root) if f.endswith('.jpeg')]

def number(nombre_archivo):
    parts = nombre_archivo.split('_')
    num = int(parts[1].split('.')[0])
    return num

lidl_images = sorted(lidl_images, key=number)
lidl_images
```

```python
all_images = lidl_images
```

#### Analitzem la inferència i creem un df

```python
data = []

for idx in range(len(all_images)):
    data_boxes = inference[idx].boxes.data

    data_boxes_np = data_boxes.cpu().detach().numpy()

    for bbx in data_boxes_np.tolist():
        data.append({
            'ticket_id': all_images[idx],
            'x1': bbx[0],
            'y1': bbx[1],
            'x2': bbx[2],
            'y2': bbx[3],
            'acc': bbx[4],
            'label': bbx[5]
        })

predicts = pd.DataFrame(data)
predicts['label_name'] = predicts['label'].map(inference[0].names)

predicts
```

```python
predicts.ticket_id.unique()
```

#### Mostrem el bbox que hem predit respecte els tiquets reals

```python
colors = {
    'import': 'red',
    'descripcio': 'blue',
    'preu_unitari': 'green',
    'quantitat_kg': 'orange',
    'total': 'black',
    'quantitat_ud': 'yellow',
}

for ticket_id, grouped_df in predicts.groupby('ticket_id'):
    pilImg = Image.open(f"{image_root}/{ticket_id}")
    print("Nom: ", ticket_id, "Size: ", pilImg.size)
    imgdrw = ImageDraw.Draw(pilImg)

    for idx, item in grouped_df.iterrows():
        imgdrw.rectangle([(item.x1, item.y1),(item.x2, item.y2)], outline = colors[item.label_name], width = 5)

    plt.imshow(pilImg)
    plt.axis('off');
    plt.show()
```
