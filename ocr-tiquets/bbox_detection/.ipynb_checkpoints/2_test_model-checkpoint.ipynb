#### Carregar model

````python
model = YOLO('./runs/detect/train_5k_15e_v3/weights/best.pt')
````

#### Inferència vàries imatges

````python
# Carpeta on es troben els tiquets que els hi volem fer inferencia.
image_root = "../../generador-tiquets/script_general/exports/data/complexity_5/coco/tickets/"

# Fem inferencia de tots els tiquets que hi ha a la carpeta.
inference = model(image_root)
````

#### Pintem el bbox predit sobre a cada imatge.


````python
for value in inference[:10]:
    pilImg = Image.open(value.path)
    imgdrw = ImageDraw.Draw(pilImg)
    dataBoxes = value.boxes.data

    for box in dataBoxes.cpu().detach().numpy():
        x = tuple(box[0:2])
        y = tuple(box[2:4])
        imgdrw.rectangle([x,y], outline = 'red', width = 2)

    plt.imshow(pilImg)
    plt.axis('off');
    plt.show()
````

#### Llegir dataset que correspon a aquestes imatges

````python
# Llegim el df dels tiquets COCO. Aqui tenim els valors reals dels tiquets.
path = '../../generador-tiquets/script_general/exports/data/complexity_5/coco/bbx_products_in_ticket.csv'
df = pd.read_csv(path, dtype = {'ticket_id': str})
df
````

#### Convertim els camps de string a tupla, ja que al guardar s'han passat a string

````python
df[['b_1', 'b_2', 'b_3', 'b_4']] = df.apply(lambda row: pd.Series([ast.literal_eval(row['b_1']), ast.literal_eval(row['b_2']), ast.literal_eval(row['b_3']), ast.literal_eval(row['b_4'])]), axis=1)
df.head(2)
````

````python
# Quantita de tiquets que tenim
len(list(df.groupby('ticket_id').groups.keys()))
````

````python
# Inferencia dels tiquets (feta previament)
inference
````

#### Creem una llista i un dataframe amb la informació de les prediccions
 ```python
[
     {
         'x1': ..,
         'y1': ..,
         'x2': ..,
         'y2': ..,
         'acc': ..,
         'label': ..
     }
 ]
```
Per crear el dataframe amb tota la informació

````python
data = []
# idx, per poder accedir a l'array de inferencia.
for idx, (ticket_id, grouped_df) in enumerate(df.groupby('ticket_id')):
    # Aquí tenim cada inferencia de cada tiquet:
    data_boxes = inference[idx].boxes.data

    # Ara agafem tots els bboxs que ha predit a dins de cada tiquet:
    # Convertim de tensor a numpy array
    data_boxes_np = data_boxes.cpu().detach().numpy()

    # Guardem en un dataframe amb la id del tiquet (aqui tenim un array de arrays on cada element és un bbox del tiquet "predicció")
    for bbx in data_boxes_np.tolist():
        data.append({
          'ticket_id': ticket_id,
          'x1': bbx[0],
          'y1': bbx[1],
          'x2': bbx[2],
          'y2': bbx[3],
          'acc': bbx[4],
          'label': bbx[5]
        })
````

````python
predicts = pd.DataFrame(data)
predicts.head()
````
