```python
from PIL import Image
import pytesseract
import numpy as np
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
import pandas as pd
import ast
```

#### Lectura del dataset amb la informació de cada part del tiquet

```python
path = '../../generador-tiquets/script_general/exports/data/complexity_5/coco/bbx_products_in_ticket.csv'
df = pd.read_csv(path, dtype = {'ticket_id': str})
df[['b_1', 'b_2', 'b_3', 'b_4']] = df.apply(lambda row: pd.Series([ast.literal_eval(row['b_1']), ast.literal_eval(row['b_2']), ast.literal_eval(row['b_3']), ast.literal_eval(row['b_4'])]), axis=1)
len(df)
```

#### Per cada registre (que serà un bbx que detecti la NN):
- Fer-ne un retall i guardar-lo en disc.
- Evaluar tesseract per cada imatge.
- Comparar-ho amb el valor que tenim guardat al df original.
- ---
- Per fer-ho:
- Agrupem el dataframe per ticket_id.
- Obrim la imatge del tiquet_id que pertoqui.
    - Dins la iteració del bucle amb agrupant el df per ticket_id.
- Per tots els elements del tiquet (l'agrupació que hem fet), fem el crop, li posem un nom i guardem.
- Posem aquest nom al df original (el que no està agrupa, el que tenia tots els elements inicialment). En aquest cas es `df`.
- Així tenim relacionat el `ground truth` (columna value) amb la imatge (columna que afegim amb el nom de la imatge que és el retall).

```python
df.index
```

```python
image_root = '../../generador-tiquets/script_general/exports/data/complexity_5/coco/tickets/'

def crop_elements(row, img):
    cropped_img = img.crop((row.b_1[0], row.b_1[1], row.b_4[0], row.b_4[1]))
    image_name = f"cut_{row.name}"
    cropped_img.save(f'./cropped_tiquets/{image_name}.png')

    return image_name

for idx, (ticket_id, grouped_df) in enumerate(df.groupby('ticket_id')):
    img = Image.open(f"{image_root}/{ticket_id}.png")

    # Crea tots els noms d'aquell tiquet en un array
    names = grouped_df.apply(lambda row: crop_elements(row, img), axis=1)

    # Afegeix aquest array a la columna cutted_img_name al ticked_id que correspon al grup
    condition = df['ticket_id'] == ticket_id
    df.loc[condition, 'cutted_img_name'] = names

df.head()
```

#### Inferència

- https://pyimagesearch.com/2021/11/15/tesseract-page-segmentation-modes-psms-explained-how-to-improve-your-ocr-accuracy/

```python
# Mostrem el progres: 
from tqdm import tqdm
pbar = tqdm(total=len(df))
```

```python
def make_inference(row):
    filename = f"./cropped_tiquets/{row.cutted_img_name}.png"
    img = np.array(Image.open(filename))
    text = pytesseract.image_to_string(img, config="--psm 7") 
    
    pbar.update()
    
    return text

df['predicted_raw_pytesseract'] = df.apply(lambda row: make_inference(row), axis = 1)

df.head()
```

```python
df['predicted_raw_pytesseract'] = df.predicted_raw_pytesseract.astype('str')
df['value'] = df.value.astype('str')
```

#### També provem de natejar les prediccions per veure si millora o no

```python
# pip install Unidecode

import string
from unidecode import unidecode

def clean(txt):
    # Signes de puntuació
    txt = txt.translate(str.maketrans('', '', string.punctuation))
    # Accents
    txt = unidecode(txt)
    # Minúscules
    txt = txt.lower()
    
    return txt
    
def tokenize(row, column_to_tokenize): 
    if row.etiqueta == 'descripcio':
        return clean(row[column_to_tokenize])
    
    return row.predicted_raw_pytesseract

df['predicted_clean_pytesseract'] = df.apply(lambda row: tokenize(row, 'predicted_raw_pytesseract'), axis = 1)
df['value_clean'] = df.apply(lambda row: tokenize(row, 'value'), axis = 1)

df.head(5)
```

#### Mirem quin tipus de etiqueta prediu millor

```python
df['predicted_raw_correct'] = (df['value'] == df['predicted_raw_pytesseract'])
df['predicted_clean_correct'] = (df['value_clean'] == df['predicted_clean_pytesseract'])

df.head()
```

#### No mirem només si està al 100% bé o malament. Examinem la tasa d'error a dins de cada imatge també.
- **CER (Character Error Rate)**: Medeix la tasa d'error en termes del percentatge de caracters que són erronis a la cadena en relació al total de caràcters. Es a dir, el percentatge de caràcters que ha fallat.
- **MER  (Mean Error Rate)**: El mateix però a nivell de paraules en comptes de caràcters.
- Es basen en: **Distancia de Levenshtein**: mesura com de semblants són dos cadenes. S'expresa amb quantes iteracions s'han de fer per passar d'una a la altre.

```python
import fastwer

# Define reference text and output text
ref = 'my name is kenneth'
output = 'myy nime iz kenneth'

# Obtain Sentence-Level Character Error Rate (CER)
print("CER: ", fastwer.score_sent(output, ref, char_level=True))

# Obtain Sentence-Level Word Error Rate (WER)
print("WER: ", fastwer.score_sent(output, ref))
```

```python
df['WER_pytesseract_raw'] = df.apply(lambda row: fastwer.score_sent(row.predicted_raw_pytesseract, row.value, char_level=True), axis = 1)
df['CER_pytesseract_raw'] = df.apply(lambda row: fastwer.score_sent(row.predicted_raw_pytesseract, row.value), axis = 1)

df['WER_pytesseract_clean'] = df.apply(lambda row: fastwer.score_sent(row.predicted_clean_pytesseract, row.value_clean, char_level=True), axis = 1)
df['CER_pytesseract_clean'] = df.apply(lambda row: fastwer.score_sent(row.predicted_clean_pytesseract, row.value_clean), axis = 1)
```

#### Mitjana dels errors de totes les vegades que hem fet inferència

```python
mean_cer_raw = df['CER_pytesseract_raw'].mean()
mean_wer_raw = df['WER_pytesseract_raw'].mean()
print(f'Mean CER = {mean_cer_raw}%, Mean WER = {mean_wer_raw}%')
```

```python
mean_cer_clean = df['CER_pytesseract_clean'].mean()
mean_wer_clean = df['WER_pytesseract_clean'].mean()
print(f'Mean CER = {mean_cer_clean}%, Mean WER = {mean_wer_clean}%')
```

```python
cer_values = [mean_cer_raw, mean_cer_clean]
wer_values = [mean_wer_raw, mean_wer_clean]

label = ['Raw', 'Clean']

x = np.arange(len(label))
bar_width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x - bar_width/2, cer_values, label='CER', color='skyblue', width=bar_width)
plt.bar(x + bar_width/2, wer_values, label='WER', color='orange', width=bar_width)

plt.ylabel('% Error')
plt.title('% Error en Raw vs Clean Data')
plt.xticks(x, label)
plt.legend()

plt.show()
```

#### Si observeme els resultats, podem dir que:
- Varia molt si natejem els càracters o símbols dels textos.
- Això ens demostra que la inferencia sense nateja no falla tant com diu, amb això ens referim a que realment no detecta com el mateix caracter perque afegeix no detecta l'accent o afegeix un . final. - - - Però el contingut real segueix tinguen sentit i és entendaor.

```python
import cv2
import easyocr

reader = easyocr.Reader(["es"], gpu=True)

def make_easyocr_inference(row):   
    filename = f"./cropped_tiquets/{row.cutted_img_name}.png"
    image = cv2.imread(filename)
    result = reader.recognize(image, paragraph=False)
    if len(result) == 0:
        return 'null'
    
    return result[0][1]
    
    
df['predicted_easyocr'] = df.apply(make_easyocr_inference, axis=1)
```

```python
# pip install keras-ocr
# pip install --force-reinstall -v "tensorflow==2.15.1"
```

```python
import keras_ocr

def make_keras_ocr_inference(row):
    filename = f"./cropped_tiquets/{row.cutted_img_name}.png"
    pipeline = keras_ocr.pipeline.Pipeline()
    
    image = [keras_ocr.tools.read(filename)]
    prediction = pipeline.recognize(image)

    llista_valors = []

    for predict in prediction[0]:
        llista_valors.append(predict[0])

    tupla_valors = tuple(llista_valors)
    resultat = ' '.join(tupla_valors)
    return resultat

df['predicted_keras_ocr'] = df.apply(make_keras_ocr_inference, axis=1)
```

#### Calcular els CER i WER per cada algoritme

```python
df['WER_easyocr'] = df.apply(lambda row: fastwer.score_sent(row.predicted_easyocr, row.value, char_level=True), axis = 1)
df['CER_easyocr'] = df.apply(lambda row: fastwer.score_sent(row.predicted_easyocr, row.value), axis = 1)

df['WER_keras_ocr'] = df.apply(lambda row: fastwer.score_sent(row.predicted_keras_ocr, row.value, char_level=True), axis = 1)
df['CER_keras_ocr'] = df.apply(lambda row: fastwer.score_sent(row.predicted_keras_ocr, row.value), axis = 1)
```

#### Comparació dels algoritmes

```python
mean_cer_pytesseract = df['CER_pytesseract_raw'].mean()
mean_wer_pytesseract = df['WER_pytesseract_raw'].mean()

mean_cer_easyocr = df['CER_easyocr'].mean()
mean_wer_easyocr = df['WER_easyocr'].mean()

mean_cer_keras_ocr = df['CER_keras_ocr'].mean()
mean_wer_keras_ocr = df['WER_keras_ocr'].mean()
```

```python
cer_values = [mean_cer_pytesseract, mean_cer_easyocr, mean_cer_keras_ocr]
wer_values = [mean_wer_pytesseract, mean_wer_easyocr, mean_wer_keras_ocr]

algoritmes = ['Pytesseract', 'EasyOCR', 'Keras_OCR']

x = np.arange(len(algoritmes))
bar_width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x - bar_width /2, cer_values, label='CER', color='skyblue', width=bar_width)
plt.bar(x + bar_width /2, wer_values, label='WER', color='orange', width=bar_width)

plt.ylabel('% Error')
plt.title('CER i WER dels Algoritmes')
plt.xticks(x, algoritmes)
plt.legend()

plt.show()
```
